```python
import librosa                                                 # Question A1
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

y, sr = librosa.load(r"C:\Users\admin\Downloads\LAB1.wav")
plt.figure(figsize=(10, 5))
librosa.display.waveshow(y)
plt.title('Speech Signal Waveform')
plt.show()
```

```python
print("Length of the signal:",len(y),"samples")                                     #Question A2
print("Sampling rate:",sr, "Hz")
print("Magnitude range:",np.min(y),"to", np.max(y))
```

```python
import IPython.display as ipd                                                        #Question A3
start = int(3 * sr)
end = int(7 * sr)
segment = y[start:end]
librosa.display.waveshow(segment)
ipd.Audio(segment, rate=sr)
```

```python
segment_1_start = int(1 * sr)                                                        #Question A4
segment_1_end = int(4 * sr)
segment_2_start = int(5 * sr)
segment_2_end = int(9 * sr)
segment_1 = y[segment_1_start:segment_1_end]
segment_2 = y[segment_2_start:segment_2_end]
print("Playing Segment 1:")

plt.figure(figsize=(10, 5))
librosa.display.waveshow(segment_1)
plt.title('Segment-1')
plt.show()
ipd.Audio(segment_1, rate=sr)
```

```python
print("Playing Segment 2:")
plt.figure(figsize=(10, 5))
librosa.display.waveshow(segment_2)
plt.title('Segment-1')
plt.show()
ipd.Audio(segment_2, rate=sr)
```
